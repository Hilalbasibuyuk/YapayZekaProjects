```python
import numpy as np
from PIL import Image
from IPython.display import Image as show_image
from keras.applications.inception_resnet_v2 import InceptionResNetV2, preprocess_input, decode_predictions
from keras.preprocessing import image
```

```python
img = Image.open("sportscar.jpg").resize((299,299))
```

```python
img = np.array(img)
```

```python
img.shape
```

```python
img.ndim # Dimension
```

```python
img = img.reshape(-1,299,299,3) # -1 ile bir boyut daha eklenir ve 4 boyutlu hale getirmi≈ü oluruz
```

```python
img.shape
```

```python
img.ndim
```

```python
img = preprocess_input(img) # scale input pixels between -1 & 1
```

```python
model = InceptionResNetV2(weights = "imagenet", classes = 1000)
```

```python
model.summary()
```

```python
type(model)
```

```python
show_image(filename = "sportscar.jpg")
```

```python
preds = model.predict(img)
print("Predicted categories:", decode_predictions(preds, top = 2)[0]) # En y√ºksek 2 kategori tahmini
```
