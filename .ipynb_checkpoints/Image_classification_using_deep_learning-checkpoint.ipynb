```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models, datasets
```

```python
(x_train, y_train), (x_test, y_test) = datasets.cifar10.load_data()
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train[:3] # Sınıflandırma indexleri
```

```python
y_test = y_test.reshape(-1,)
y_test # Tek bir boyuta indirgedik
```

```python
image_class = ["airplane", "automobile", "bird", "cat", "deer", "dog", "frog", "horse", "ship", "truck"]
```

```python
def plot_sample(x,y,index):
    plt.figure(figsize = (15,2))
    plt.axis("off")
    plt.imshow(x[index])
    plt.title(image_class[y[index]])
```

```python
plot_sample(x_test,y_test,0)
```

```python
plot_sample(x_test,y_test,1)
```

```python
plot_sample(x_test,y_test,4)
```

```python
plot_sample(x_test,y_test,9)
```

```python
plot_sample(x_test,y_test,3)
```

```python
# Normalization
x_train = x_train/255
x_test = x_test/255
```

```python
model = models.Sequential([
    layers.Conv2D(filters = 32, kernel_size=(3,3), activation = "relu", input_shape= (32,32,3)),
    layers.MaxPooling2D((2,2)),
    
    layers.Conv2D(filters = 64, kernel_size = (3,3), activation = "relu"),
    layers.MaxPooling2D((2,2)),
    
    layers.Flatten(),
    layers.Dense(64, activation = "relu"),
    layers.Dense(10, activation = "softmax") # 10 class
])
```

```python
model.compile(optimizer = "adam",
             loss = "sparse_categorical_crossentropy",
             metrics = ["accuracy"])
```

```python
model.fit(x_train,y_train,epochs = 5)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pred = model.predict(x_test)
y_pred[:3]
```

```python
y_predict_classes = [np.argmax(element) for element in y_pred]
y_predict_classes[:3]
```
