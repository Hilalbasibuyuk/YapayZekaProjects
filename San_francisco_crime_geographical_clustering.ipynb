## 1- Import libraries and prepare the dataset

```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv("train.csv")
data.head(7)
```

```python
data = data.drop(["Descript", "DayOfWeek", "PdDistrict", "Resolution", "Address"], axis = 1)
data.head(7)
```

```python
data.isnull().sum() # Nan değer yok
```

### This dataset contains information from 2003 to 2015. We will use only data in year 2014

```python
# Dates'ten saat kısmını çıkarma
del1 = lambda x: (x["Dates"].split())[0]
data["Dates"] = data.apply(del1, axis =1)
data.head(7)
```

```python
# Tarih kısmı hariç çıkarma işlemi
del2 = lambda x: (x["Dates"].split("-"))[0]
data["Dates"] = data.apply(del2, axis =1)
data.tail(7) # En alttaki veriler
```

```python
data_2014 = data[(data.Dates == "2014")] # Sadece 2014 yılındaki verileri aldık
data_2014.head(7)
```

```python
scaler = MinMaxScaler()

scaler.fit(data_2014[["X"]])
data_2014["X_scaled"] = scaler.transform(data_2014[["X"]])

scaler.fit(data_2014[["Y"]])
data_2014["Y_scaled"] = scaler.transform(data_2014[["Y"]])
```
